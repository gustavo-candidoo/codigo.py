```python
# Exercise 59: Get Ibovespa quotes since June 1, 1994 through Yahoo Finance.
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
ibv = "^BVSP" 

df_ibv = yf.download(tickers=ibv, start="1994-06-01", end=datetime.now())

df_ibv
```
